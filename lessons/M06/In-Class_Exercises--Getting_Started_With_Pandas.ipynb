```python
# 3
iris = sns.load_dataset('iris')
iris.sort_values(by = ['species'], ascending = False)
```

```python
# 4
iris['species'].value_counts().to_frame('value count')
```

```python
# 5
len(iris[(iris['petal_length'] <= 1.4) | (iris['petal_length'] >= 1.6)])
iris[(iris['petal_length'] <= 1.4) | (iris['petal_length'] >= 1.6)].shape[0]
```

```python
# 6
iris['min_petal'] = iris.apply(lambda x: min(x['petal_length'], x['petal_width']), axis = 1)
iris
```

```python
# 6
print(sum(iris.petal_length <= iris.petal_width))
iris[['petal_length', 'petal_width']].idxmin(axis = 1).value_counts()
```

```python
# 7
pd.pivot_table(iris, values = ['sepal_width', 'petal_width'], columns = ['species'], aggfunc = {np.mean, np.median})
```
